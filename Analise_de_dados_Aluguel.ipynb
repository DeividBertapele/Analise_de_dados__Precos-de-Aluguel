# <font color= '#03fc73'> DEFINIÇÃO DE MODELOS: </font>

<font color= '#03fc73'> COM MACHINE LEARNING -> SCIKIT-LEARN => </font> <font color='blue'> https://scikit-learn.org/stable/ </font>

<font color='blue'> https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html?highlight=train_test_split#sklearn.model_selection.train_test_split </font>


```python
# Vamos separar as nossas variáveis preditoras 

X = base.drop('aluguel', axis=1)
y = base.aluguel
```

```python
# Importando o train_test_split

from sklearn.model_selection import train_test_split
```

```python
# Separando a nossa base

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)  
```

```python
# Visualizando X_train

X_train.head(5)
```

## <font color= '#03fc73'> REGRESSÃO LINEAR : </font> 

<font color='blue'> https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html?highlight=linearregression#sklearn.linear_model.LinearRegression </font>

```python
# Importando Regressão linear

from sklearn.linear_model import LinearRegression
```

```python
# Mostrando o modelo o que você quiser

reg = LinearRegression().fit(X_train, y_train)
```

```python
# Avaliando o ajuste do modelo aos nosso dados de TREINO

reg.score(X_train, y_train)
```

<font color= '#03fc73'> REGRESSÃO MÉTRICAS : 

<font color= 'red'> link: https://scikit-learn.org/stable/modules/classes.html?highlight=metrics#module-sklearn.metrics </font> 

```python
# Importando o erro médio absoluto

from sklearn.metrics import mean_absolute_error
```

```python
# Fazendo previsão com modelo anterior

y_pred = reg.predict(X_test)
```

```python
# Avaliando o erro

mean_absolute_error(y_test, y_pred)
```

```python
# Visualizando essa diferença graficamente

fig, ax = plt.subplots(figsize=(9,9))

ax.scatter(y_test, y_pred)
ax.plot([0, 10000], [0, 10000])

plt.show()
```

<font color= '#03fc73'> ÁRVORE DE DECISÃO : 

<font color= 'red'> link: https://scikit-learn.org/stable/modules/tree.html#regression  </font> 

```python
# Importando a árvore de decisão

from sklearn import tree
```

```python
# Criando o regressor

regArvore = tree.DecisionTreeRegressor(random_state=42)
```

```python
# Fazendo o fit do modelo

regArvore = regArvore.fit(X_train, y_train)
```

```python
# Avaliando o modelo 

regArvore.score(X_train, y_train)
```

```python
# Fazendo a previsão

y_predArvore = regArvore.predict(X_test)
```

```python
# Avaliando o erro para árvore de decisão

mean_absolute_error(y_test, y_predArvore)
```

```python
# Visualizando essa diferença graficamente

fig, ax = plt.subplots(figsize=(9,9))

ax.scatter(y_test, y_pred)
ax.scatter(y_test, y_predArvore)
ax.plot([0, 10000], [0, 10000])

plt.show()
```
